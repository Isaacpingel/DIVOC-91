### Clean data to make it consistent with other survey data for analysis

```python
# Updates to 2023 2-4q dataframes
# Function to Collapse WHENCOVIDRV1, WHENCOVIDRV2, and WHENCOVIDRV3 column responses into a single column for analysis
def set_whencovid_2q_clean(row):
    """Collapse WHENCOVIDRV1, WHENCOVIDRV2, and WHENCOVIDRV3 column values into a single column"""

    #In case of multiple responses, the most recent diagnosis will be used 
    if row['WHENCOVIDRV1'] == 1:
        whencovid_clean = 1
    elif row['WHENCOVIDRV2'] == 1:
        whencovid_clean = 2
    elif row['WHENCOVIDRV3'] == 1:
        whencovid_clean = 3
    else:
        whencovid_clean = 99
        
    return whencovid_clean

# Create new WHENCOVID_CLEAN column and load to DataFrame using set_whencovid_2q_clean function
pulse2023_2q_analysis_wrk_df["WHENCOVID_CLEAN"] = pulse2023_2q_analysis_wrk_df.apply(set_whencovid_2q_clean, axis=1)
pulse2023_3q_analysis_wrk_df["WHENCOVID_CLEAN"] = pulse2023_3q_analysis_wrk_df.apply(set_whencovid_2q_clean, axis=1)
pulse2023_4q_analysis_wrk_df["WHENCOVID_CLEAN"] = pulse2023_4q_analysis_wrk_df.apply(set_whencovid_2q_clean, axis=1)

display(pulse2023_2q_analysis_wrk_df.head())
display(pulse2023_3q_analysis_wrk_df.head())
display(pulse2023_4q_analysis_wrk_df.head())
```

```python
# Updates to 2023 1q dataframe
# Function to change -88 and -99 (No Response) to common response of 99 for analysis
def clean_negative_values(response):
    """Change -88 and -99 (No Response) to common response of 99"""

    if response in [-88, -99]:
        return 99
    else:
        return response

# Create new WHENCOVID_CLEAN column and load to DataFrame using clean_negative_values function
pulse2023_1q_analysis_wrk_df["WHENCOVID_CLEAN"] = pulse2023_1q_analysis_wrk_df["WHENCOVID"].apply(clean_negative_values)

display(pulse2023_1q_analysis_wrk_df.head())
```

```python
# Add survey start and end date columns to quarterly surveys
# 2023-1Q
pulse2023_1q_analysis_wrk_df['SURVEY_START_DATE'] = pd.to_datetime("2023-03-01")
pulse2023_1q_analysis_wrk_df['SURVEY_END_DATE'] = pd.to_datetime("2023-03-13")
# 2023-2Q
pulse2023_2q_analysis_wrk_df['SURVEY_START_DATE'] = pd.to_datetime("2023-06-07")
pulse2023_2q_analysis_wrk_df['SURVEY_END_DATE'] = pd.to_datetime("2023-06-19")
# 2023-3Q
pulse2023_3q_analysis_wrk_df['SURVEY_START_DATE'] = pd.to_datetime("2023-07-26")
pulse2023_3q_analysis_wrk_df['SURVEY_END_DATE'] = pd.to_datetime("2023-08-07")
# 2023-4Q
pulse2023_4q_analysis_wrk_df['SURVEY_START_DATE'] = pd.to_datetime("2023-10-18")
pulse2023_4q_analysis_wrk_df['SURVEY_END_DATE'] = pd.to_datetime("2023-10-30")
```
