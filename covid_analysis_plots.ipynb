```python
# Import Libraries and Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Prepare Data For Analysis

```python
# Read the US Census Bureau Household Pulse Survey CSV files into DataFrames.
pulse2021_survey_data_file = 'Resources/pulse2021_survey_data.csv'
pulse2021_survey_data_df = pd.read_csv(pulse2021_survey_data_file,
#                                      index_col='SURVEY_END_DATE',
                                       parse_dates=True)

pulse2022_survey_data_file = 'Resources/pulse2022_survey_data.csv'
pulse2022_survey_data_df = pd.read_csv(pulse2022_survey_data_file,
#                                      index_col='SURVEY_END_DATE',
                                       parse_dates=True)

pulse2023_survey_data_file = 'Resources/pulse2023_survey_data.csv'
pulse2023_survey_data_df = pd.read_csv(pulse2023_survey_data_file,
#                                      index_col='SURVEY_END_DATE',
                                       parse_dates=True)

print("Input files have been read in")
```

```python
#Display first 5 rows of dataframes
display(pulse2021_survey_data_df.head())
display(pulse2022_survey_data_df.head())
display(pulse2023_survey_data_df.head())
```

```python
#Get shape of data files
display(pulse2021_survey_data_df.shape)
display(pulse2022_survey_data_df.shape)
display(pulse2023_survey_data_df.shape)
```

```python
#Get information about data fields
display(pulse2021_survey_data_df.info())
display(pulse2022_survey_data_df.info())
display(pulse2023_survey_data_df.info())
```

```python
# Concatenate survey files for analysis
pulse_survey_data_df = pd.concat([pulse2021_survey_data_df, pulse2022_survey_data_df,
                                       pulse2023_survey_data_df], ignore_index=True)
```

```python
#Get shape of concatenated file
display(pulse_survey_data_df.shape)
```

```python
# Convert Survey dates to datetime fields
pulse_survey_data_df['SURVEY_START_DATE'] = pd.to_datetime(pulse_survey_data_df['SURVEY_START_DATE'], errors='coerce')
pulse_survey_data_df['SURVEY_END_DATE'] = pd.to_datetime(pulse_survey_data_df['SURVEY_END_DATE'], errors='coerce')
```

```python
# Get information about data fields after datetime data type changes
display(pulse_survey_data_df['SURVEY_START_DATE'].info())
display(pulse_survey_data_df['SURVEY_END_DATE'].info())
```

```python
# Set dataframe index to SURVEY_END_DATE
pulse_survey_data_df.set_index('SURVEY_END_DATE', inplace=True)
```

```python
# Display first and last 5 rows
display(pulse_survey_data_df.head())
display(pulse_survey_data_df.tail())
```

```python
#Restore point:  Make copy of dataframe before changes
pulse_analysis_clean_df = pulse_survey_data_df.copy()
pulse_analysis_clean_df.head()
print("pulse_survey_data_df -- Restore point created")
```

```python
# Convert multiple survey responses into a single response for analysis

# Responses:
#  1 - Not at all / No - no difficulty
#  2 - Several days / Yes - some difficulty
#  3 - More than half the days / Yes - a lot of difficulty
#  4 - Nearly every day / Cannot do at all

#  Responses 2-4 will be combined into an analysis response of 5 to reflect that the respondent is experiencing this at
#  level.  This will then be compared to respondents that are not experiencing this at all.

def combine_responses(response):
    """Convert 2, 3, and 4 responses to a combined value of 5"""
    if response in [2, 3, 4]:
        return 5
    else:
        return response

# List of columns that will be cleaned
clean_columns_list = ['ANXIOUS', 'WORRY', 'INTEREST', 'DOWN', 'SEEING', 'HEARING', 'MOBILITY', 'REMEMBERING',
                      'SELFCARE', 'UNDERSTAND']

# Create new '..._CLEAN' columns and load to DataFrame using combine_responses function
for col in clean_columns_list:
    pulse_analysis_clean_df[col + '_CLEAN'] = pulse_analysis_clean_df[col].apply(combine_responses)

pulse_analysis_clean_df.head()
```

# Plot 1:  COVID and Anxiety

```python
#Create dataframe of respondents by HAD_COVID and ANXIOUS_CLEAN
hadcovid_anxious_groupby_df = pulse_analysis_clean_df.groupby(['SURVEY_END_DATE', 'HAD_COVID',
                                                               'ANXIOUS_CLEAN']).size().reset_index(name='count')
```

```python
#display dataframe
display(hadcovid_anxious_groupby_df)
```

```python
hadcovid_anxious_groupby_df.info()
```

```python
#Remove the No Response answers
hadcovid_anxious_groupby_df = hadcovid_anxious_groupby_df.loc[hadcovid_anxious_groupby_df['HAD_COVID'] != 99]
hadcovid_anxious_groupby_df = hadcovid_anxious_groupby_df.loc[hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'] != 99]
hadcovid_anxious_groupby_df.head()
```

```python
#Get count of rows and columns
display(hadcovid_anxious_groupby_df.shape)
display(hadcovid_anxious_groupby_df['SURVEY_END_DATE'].value_counts())
display(hadcovid_anxious_groupby_df['HAD_COVID'].value_counts())
display(hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'].value_counts())
```

```python
#Create a dataframe of respondents that had COVID and have no Anxiety
hadcovid1_anxious1_groupby_df = hadcovid_anxious_groupby_df.loc[(hadcovid_anxious_groupby_df['HAD_COVID'] == 1) &
                                                                (hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'] == 1)]

hadcovid1_anxious1_groupby_df.set_index('SURVEY_END_DATE', inplace=True)
display(hadcovid1_anxious1_groupby_df.head())
display(hadcovid1_anxious1_groupby_df.shape)
```

```python
#Create a dataframe of respondents that had COVID and have some level of Anxiety
hadcovid1_anxious5_groupby_df = hadcovid_anxious_groupby_df.loc[(hadcovid_anxious_groupby_df['HAD_COVID'] == 1) &
                                                                (hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'] == 5)]

hadcovid1_anxious5_groupby_df.set_index('SURVEY_END_DATE', inplace=True)
display(hadcovid1_anxious5_groupby_df.head())
display(hadcovid1_anxious5_groupby_df.shape)
```

```python
#Create a dataframe of respondents that have not had COVID and have no Anxiety
hadcovid2_anxious1_groupby_df = hadcovid_anxious_groupby_df.loc[(hadcovid_anxious_groupby_df['HAD_COVID'] == 2) &
                                                                (hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'] == 1)]

hadcovid2_anxious1_groupby_df.set_index('SURVEY_END_DATE', inplace=True)
display(hadcovid2_anxious1_groupby_df.head())
display(hadcovid2_anxious1_groupby_df.shape)
```

```python
#Create a dataframe of respondents that have not had COVID and have some level of Anxiety
hadcovid2_anxious5_groupby_df = hadcovid_anxious_groupby_df.loc[(hadcovid_anxious_groupby_df['HAD_COVID'] == 2) &
                                                                (hadcovid_anxious_groupby_df['ANXIOUS_CLEAN'] == 5)]

hadcovid2_anxious5_groupby_df.set_index('SURVEY_END_DATE', inplace=True)
display(hadcovid2_anxious5_groupby_df.head())
display(hadcovid2_anxious5_groupby_df.shape)
```

```python
# Visualizing responses HAD_COVID and ANXIOUS_CLEAN
  
# Set the plot size 
plt.figure(figsize=(16, 8), dpi=150) 
  
# Plot responses and set the label and color of the line 
hadcovid1_anxious1_groupby_df['count'].plot(label='Had COVID, No Anxiety', color='orange')
hadcovid1_anxious5_groupby_df['count'].plot(label='Had COVID, Some Anxiety', color='blue')
hadcovid2_anxious1_groupby_df['count'].plot(label='No COVID, No Anxiety', color='red')
hadcovid2_anxious5_groupby_df['count'].plot(label='No COVID, Some Anxiety', color='green')
  
# Add title to the plot 
plt.title('COVID and Anxiety Plot') 
  
# Add label to the x-axis 
plt.xlabel('Quarters') 
  
# Add legend 
plt.legend()

# show the plot 
plt.show()
```
