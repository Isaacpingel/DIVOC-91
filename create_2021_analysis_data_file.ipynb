### Clean data to make it consistent with other survey data for analysis

```python
# Define clean_negative_values function to change -88 and -99 (No Response) to common response of 99 for analysis
def clean_negative_values(response):
    """Change -88 and -99 (No Response) to common response of 99"""

    if response in [-88, -99]:
        return 99
    else:
        return response

print("clean_negative_values function defined")
```

```python
# Get list of 1q file columns to identify which ones need the responses cleaned
pulse2021_1q_analysis_wrk_df.columns
```

```python
# 1q file cleanup:  Replace all negative no response values (-88 and -99) with common value of 99 in dataframe
#                   using clean_negative_values function
clean_columns_list = ['TBIRTH_YEAR', 'RHISPANIC', 'RRACE', 'EEDUC', 'MS', 'THHLD_NUMPER', 'RECVDVACC',
                      'HADCOVID', 'ANXIOUS', 'WORRY', 'INTEREST', 'DOWN', 'ANYWORK', 'KINDWORK', 'EXPNS_DIF',
                      'CURFOODSUF', 'FREEFOOD', 'TENURE', 'PRIVHLTH', 'PUBHLTH', 'INCOME', 'EGENDER']

for col in clean_columns_list:
    pulse2021_1q_analysis_wrk_df[col] = pulse2021_1q_analysis_wrk_df[col].apply(clean_negative_values)
    
# Confirm that no -88 or -99 values are in the dataframe
print("Checking to confirm that pulse2021_1q_analysis_wrk_df does not contain any -88 or -99 values")
negative_values_list = [-88, -99]
for response in negative_values_list:
    if response in pulse2021_1q_analysis_wrk_df.values:
        print(f"{response} exists in dataframe")
    else:
        print(f"{response} does not exist in dataframe")
```
