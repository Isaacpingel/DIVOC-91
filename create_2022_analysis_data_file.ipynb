### Clean data to make it consistent with other survey data for analysis

```python
# Clean WHENCOVID columns and create new WHENCOVID_CLEAN column

# Function to change -88 and -99 (No Response) to common response of 99 for analysis
def clean_negative_values(response):
    """Change -88 and -99 (No Response) to common response of 99"""

    if response in [-88, -99]:
        return 99
    else:
        return response

# Create new WHENCOVID_CLEAN column and load to DataFrame using clean_negative_values function for 3q and 4q files
# Set 1q and 2q file responses to 99 (No Response) because data was not collected at that time
pulse2022_1q_analysis_wrk_df['WHENCOVID_CLEAN'] = 99
pulse2022_2q_analysis_wrk_df['WHENCOVID_CLEAN'] = 99
pulse2022_3q_analysis_wrk_df['WHENCOVID_CLEAN'] = pulse2022_3q_analysis_wrk_df['WHENCOVID'].apply(clean_negative_values)
pulse2022_4q_analysis_wrk_df['WHENCOVID_CLEAN'] = pulse2022_4q_analysis_wrk_df['WHENCOVID'].apply(clean_negative_values)

display(pulse2022_1q_analysis_wrk_df.head())
display(pulse2022_2q_analysis_wrk_df.head())
display(pulse2022_3q_analysis_wrk_df.head())
display(pulse2022_4q_analysis_wrk_df.head())
```
